```python
# Declaring the below custom transformer for sklearn 
# for showing effects like carry over and saturation simultaneously on the dataset

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.utils.validation import check_is_fitted, check_array

class ExponentialSaturation(BaseEstimator, TransformerMixin):
    def __init__(self, a=1.):
        self.a = a
        
    def fit(self, X, y=None):
        X = check_array(X)
        self._check_n_features(X, reset=True) # from BaseEstimator
        return self

    def transform(self, X):
        check_is_fitted(self)
        X = check_array(X)
        self._check_n_features(X, reset=False) # from BaseEstimator
        return 1 - np.exp(-self.a*X)
    


class ExponentialCarryover(BaseEstimator, TransformerMixin):
    def __init__(self, strength=0.5, length=1):
        self.strength = strength
        self.length = length

    def fit(self, X, y=None):
        X = check_array(X)
        self._check_n_features(X, reset=True)
        self.sliding_window_ = (
            self.strength ** np.arange(self.length + 1)
        ).reshape(-1, 1)
        return self

    def transform(self, X: np.ndarray):
        check_is_fitted(self)
        X = check_array(X)
        self._check_n_features(X, reset=False)
        convolution = convolve2d(X, self.sliding_window_)
        if self.length > 0:
            convolution = convolution[: -self.length]
        return convolution
```

```python
adstock = ColumnTransformer(
    [
     ('sales_rep_calls_pipe', Pipeline([
                           ('carryover', ExponentialCarryover()),
                           ('saturation', ExponentialSaturation())
     ]), ['Sales_Rep_Calls']),
     ('samples_dropped_pipe', Pipeline([
                           ('carryover', ExponentialCarryover()),
                           ('saturation', ExponentialSaturation())
     ]), ['Samples_Dropped']),
     ('emails_delivered_pipe', Pipeline([
                           ('carryover', ExponentialCarryover()),
                           ('saturation', ExponentialSaturation())
     ]), ['Emails_Delivered']),
    ],
    remainder='passthrough'
)

model = Pipeline([
                  ('adstock', adstock),
                  ('regression', LinearRegression())
])


def findbestfea(ssrc, ssd, sed, slsrc, slsd, sled):
    _roi1 = slsrc / (ssrc+1)
    _roi2 = slsd / (ssd+1)
    _roi3 = sled / (sed+1)
    if(_roi1> max(_roi2, _roi3)):
        return 0
    elif _roi2> max(_roi1, _roi3):
        return 1
    else :
        return 2



weightcolumns =  ['Base','Sales_Rep_Calls', 'Samples_Dropped',  'Emails_Delivered', 'Specialty', 'Physician_Segment', 'Market_Rx', 'Speaker_Programs_Attended', 
            'Vouchers_Dropped']

cntri = list()
contri = dict()
```

```python
j  =0
for phyid in phyId:
    j+=1
    df = pd.DataFrame(data[data['Physician_ID'] == phyid])
    df.set_index('Time_Period', inplace = True)
    dx =  df[df['Physician_ID'] == phyid].drop(columns=['Brand_Rx','Physician_ID'])
    y =  df[df['Physician_ID'] == phyid]
    y = y['Brand_Rx']
    X = dx
    if(y.sum() == 0):
        y.replace(to_replace = 1e-9, inplace =  True)
    model.fit(X, y)
    
#     print(cross_val_score(model, X, y, cv=TimeSeriesSplit()).mean())
#   The above line is showing some error for some particular observation so commented, I haven't much researched

    tuned_model = OptunaSearchCV(
    estimator=model,
    param_distributions={
        'adstock__sales_rep_calls_pipe__carryover__strength': UniformDistribution(0, 1),
        'adstock__sales_rep_calls_pipe__carryover__length': IntUniformDistribution(0, 6),
        'adstock__sales_rep_calls_pipe__saturation__a': UniformDistribution(0, 0.01),
        'adstock__samples_dropped_pipe__carryover__strength': UniformDistribution(0, 1),
        'adstock__samples_dropped_pipe__carryover__length': IntUniformDistribution(0, 6),
        'adstock__samples_dropped_pipe__saturation__a': UniformDistribution(0, 0.01),
        'adstock__emails_delivered_pipe__carryover__strength': UniformDistribution(0, 1),
        'adstock__emails_delivered_pipe__carryover__length': IntUniformDistribution(0, 6),
        'adstock__emails_delivered_pipe__saturation__a': UniformDistribution(0, 0.01),
    },
    n_trials=1000,  # You can increase the n_trials to upto 1000 but my machine is taking too much time(greater than 2 hour) 
#   for running on the 10 trial only so decide as per your system.
    cv=TimeSeriesSplit(),
    random_state=0
    )
    tuned_model.fit(X, y)
    
# The optimal hyperparameters are the following:
#     print(tuned_model.best_params_)
#     print(tuned_model.best_estimator_.named_steps['regression'].coef_)
#     print(tuned_model.best_estimator_.named_steps['regression'].intercept_)

    
    adstock_data = pd.DataFrame(
    tuned_model.best_estimator_.named_steps['adstock'].transform(X),
    columns=X.columns,
    index=X.index
    )


    weight = pd.Series(
    tuned_model.best_estimator_.named_steps['regression'].coef_,
    index=X.columns
    )
    base = tuned_model.best_estimator_.named_steps['regression'].intercept_

    unadj_contributions = adstock_data.mul(weight).assign(Base=base)  #(unadjusted contribution)
    adj_contributions = (unadj_contributions
                     .div(unadj_contributions.sum(axis=1), axis=0)   #(adjusted contribution)
                     .mul(y, axis=0)
                    )
    slsrc = adj_contributions['Sales_Rep_Calls'].sum()
    slsd = adj_contributions['Samples_Dropped'].sum()
    sled = adj_contributions['Emails_Delivered'].sum()
    ssrc = X['Sales_Rep_Calls'].sum()
    ssd = X['Samples_Dropped'].sum()
    sed = X['Emails_Delivered'].sum()
    cntri.append(findbestfea(ssrc, ssd, sed, slsrc, slsd, sled))
    contri[phyid] = findbestfea(ssrc, ssd, sed, slsrc, slsd, sled)
    X_new = pd.DataFrame({
    'Sales_Rep_Calls': [0, 0, 0],
    'Samples_Dropped': [0, 3000, 0],
    'Emails_Delivered': [1000, 1000, 1000]
    })
    tuned_model.predict(X_new)

#     Getting some error for plotting the graph of 13th phsician graph (as per the phyId array) so commented below code
#     ax = (adj_contributions[weightcolumns]
#       .plot.area(
#           figsize=(16, 10),
#           linewidth=1,
#           title='Predicted Sales and Breakdown',
#           ylabel='Sales',
#           xlabel='Date'
#       )
#      )
    
#     handles, labels = ax.get_legend_handles_labels()
#     ax.legend(
#     handles[::-1], labels[::-1],
#     title='Channels', loc="center left",
#     bbox_to_anchor=(1.01, 0.5)
#     )
```
